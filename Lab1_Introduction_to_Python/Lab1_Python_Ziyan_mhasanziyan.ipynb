container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

```python
# The error 'TypeError: 'tuple' object does not support item assignment' occurs because
# tuples in Python are immutable. This means that once a tuple is created, its elements
# cannot be changed. The operation `tup[2] = 23` attempts to modify an element,
# which is not allowed for a tuple.
#
# If you need to 'modify' a tuple, you must create a new tuple with the desired changes.
# For example, to change the element at index 2 to 23, you can do the following:
tup = tup[:2] + (23,) + tup[3:]
# Now, the 'tup' variable refers to a new tuple with the updated value at index 2.
# If you print tup now, you will see the change.
# print(tup)
```

```python
# Problem 1: Confusion Matrix Metrics
TP = 50
FP = 8
TN = 32
FN = 10

total = TP + FP + TN + FN
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)


```

```python
SSR = 420
SSE = 180

SST = SSR + SSE


r_squared = SSR / SST


print("Total Sum of Squares (SST):", SST)
print("R-squared:", r_squared)


if r_squared > 0.7:
    print("The model explains more than 70% of the variance.")
else:
    print("The model explains less than 70% of the variance.")
```

```python
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

mse = (e1 + e2 + e3 + e4 + e5) / 5
print("Mean Squared Error:", mse)
```

```python
ei_sq = 9
MSE = 3
p = 3
hii = 0.2
n = 25

cook_d = (ei_sq / (p * MSE)) * (hii / (1 - hii)**2)
threshold = 4 / n

print("Cook's Distance:", cook_d)
print("Threshold:", threshold)

if cook_d > threshold:
    print("Influential observation")
else:
    print("Not influential")
```
